## Caching
In computing, a cache is a high-speed data storage layer which stores a subset of data, typically transient in nature, so that future requests for that data are served up faster than is possible by accessing the data’s primary storage location. Caching allows you to efficiently reuse previously retrieved or computed data.



Auto-regressive text generation is the process used by LLMs to generate text in a sequential manner, one token at a time.
It works by iteratively:
   - Taking an input sequence of tokens
   - Sampling the next token based on the model's predictions
   - Appending the sampled token to the input sequence
   - Repeating the process until a stop token is generated
This process is called "auto-regressive" because the model's predictions for the next token depend on the previously generated tokens, forming a self-referential or "regressive" process.


LLMs undergo training using the causal language modeling objective, which necessitates masking the upper triangular portion of the attention score matrix. 
This ensures that tokens never depend on future tokens during training.
